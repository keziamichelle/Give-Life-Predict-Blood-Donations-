```python
import pandas as pd
data=pd.read_csv("C:/Users/ASUS/Desktop/transfusion.csv")
data.head()
```

```python
from tpot import TPOTClassifier
```

```python
data.rename(columns={"whether he/she donated blood in March 2007": "target"}, inplace=True)
data.head()
```

```python
data.target.value_counts(normalize=True)
```

```python
X=data.drop('target',axis=1)
X
```

```python
Y=data['target']
Y
```

```python
from imblearn.over_sampling import SMOTE
o=SMOTE(random_state=42)
X,Y=o.fit_resample(X,Y)
X
```

```python
Z=pd.concat([X,Y],axis=1)
Z
```

```python
Z['target'].value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25, random_state=42)
xtrain.head()
```

```python
from sklearn.metrics import roc_auc_score
```

```python
tpot=TPOTClassifier(generations=5, population_size=20, verbosity=2, scoring='roc_auc', random_state=42, disable_update_check=True, config_dict='TPOT light')
tpot.fit(xtrain,ytrain)
```

```python
tpot_auc_score=roc_auc_score(ytest, tpot.predict_proba(xtest)[:,1])
print(f'\nAUC score: {tpot_auc_score:.4f}')
```

```python
print("\nThe best pipeline: ", )
for idx, (name, transform) in enumerate(tpot.fitted_pipeline_.steps, start=1):
    print(f'{idx}. {transform}')
```
